## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Primary Keys are the unique identifier of a record within a table.  

A foreign key is a key contained within table A, that is the primary key in another table (B).  It may not be unique in table A, but it allows one to JOIN with table B to retrieve additional data.  It will be unique in table B.    



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

When considering the relationships between different tables within a database, they can be of the following types:

One-to-One: Each record in table A can be mapped to one record in table B (e.g. Table of Names and Table of Birthdays)

One-to-Many: Each record in table A can be mapped to multiple records in table B (e.g. Table of Addresses, and Table of Names.  More than one person may live at each address).

Many-to-Many: Multiple records in table A map to multiple records in table B (e.g. table of cities, and table of restaurant brands).

LEFT JOIN: Takes all records from the left table, but only matching records from the right table are shown.

RIGHT JOIN: Takes all records from the right table, but only matching records from the left table are shown.

INNER JOIN: Only records that appear in both tables are displayed.

OUTER JOIN: Only records that appear in neither table are displayed.
